### 3. Rotasi

```python
def rotate_image(image, angle):
    angle = angle % 360
    h, w = image.shape

    if angle == 90:
        return np.rot90(image, k=1)
    elif angle == 180:
        return np.rot90(image, k=2)
    elif angle == 270:
        return np.rot90(image, k=3)
    else:
        return image

rotate1 = rotate_image(image1, 270)
plt.imshow(rotate1, cmap='gray')
plt.title("Citra 1")  
plt.show()

rotate2 = rotate_image(image2, 90)
plt.imshow(rotate2, cmap='gray')
plt.title("Citra 2")  
plt.show()

rotate3 = rotate_image(image3, 180)
plt.imshow(rotate3, cmap='gray')
plt.title("Citra 3")  
plt.show()
```

rotate_image() menerima gambar dan sudut, lalu memastikan sudutnya tetap dalam rentang 0–359 dengan operasi modulo. Setelah itu, fungsi mengecek sudutnya dan menggunakan np.rot90() untuk memutar gambar sesuai jumlah rotasi yang dibutuhkan. Kalau sudutnya bukan salah satu dari tiga itu, ya gambar dikembalikan apa adanya

### 4. Flip Citra

```python
def flip(image, mode='X'):
    height, width = image.shape                       
    result = np.zeros((height, width), dtype=image.dtype)
    
    for i in range(height):                      
        for j in range(width):                  
            if mode == 'X':
                result[i, j] = image[i, width-1-j]
            elif mode == 'Y':
                result[i, j] = image[height-1-i, j]
    
    return result

flip_4 = flip(image4, 'X')
flip_3 = flip(rotate3, 'X')

plt.subplot(1, 2, 1)
plt.imshow(flip_3, cmap='gray')
plt.title("Citra 3")

plt.subplot(1, 2, 2)
plt.imshow(flip_4, cmap='gray')
plt.title("Citra 4")

plt.tight_layout()
plt.show()
```

Fungsi flip() menerima gambar dan mode. Mode 'X' berarti membalik gambar secara horizontal (kiri-kanan), sedangkan 'Y' untuk vertikal (atas-bawah). Kode membuat array kosong dengan ukuran sama, lalu mengisi ulang piksel satu per satu menggunakan dua loop, dengan indeks dibalik sesuai arah flip yang dipilih.

```python

fig, axes = plt.subplots(2, 2, figsize=(10, 10))

axes[0, 0].imshow(rotate1, cmap='gray')
axes[0, 0].set_title("Citra 1")

axes[0, 1].imshow(flip_3, cmap='gray')
axes[0, 1].set_title("Citra 2")

axes[1, 0].imshow(rotate2, cmap='gray')
axes[1, 0].set_title("Citra 3")

axes[1, 1].imshow(flip_4, cmap='gray')
axes[1, 1].set_title("Citra 4")

plt.tight_layout()
plt.show()
```

### 6. penggabungan

```python
h, w = crop1.shape

kanvas = np.zeros((h*2, w*2))

kanvas[0:h, 0:w] = crop1
kanvas[0:h, w:2*w] = crop2
kanvas[h:2*h, 0:w] = crop3
kanvas[h:2*h, w:2*w] = crop4


kanvasBagi2 = (kanvas.astype(np.float32) / 2)
plt.imshow(kanvasBagi2, cmap='gray')
plt.title("Hasil Gabungan Buggy")
plt.axis('off')
plt.show()
```

### 7. histogram

```python
def hitung_histogram(image):
    histogram = np.zeros(256, dtype=int)
    for pixel in image.flatten():
        histogram[int(pixel)] += 1
    return histogram

def normalize(image):
    min_val = image.min()
    max_val = image.max()
    normalized = ((image - min_val) / (max_val - min_val) * 255).astype(np.uint8)
    return normalized

hist_awal = hitung_histogram(kanvasBagi2)
kanvas_normalized = normalize(kanvasBagi2)
hist_normal = hitung_histogram(kanvas_normalized)


plt.subplot(2, 2, 3)
plt.bar(range(256), hist_awal, color='blue', width=1)
plt.title('Histogram')
plt.xlabel('Nilai Intensitas')
plt.ylabel('Jumlah Piksel')
plt.xlim([0, 255])

plt.subplot(2, 2, 4)
plt.bar(range(256), hist_normal, color='green', width=1)
plt.title('Histogram Setelah Normalisasi')
plt.xlabel('Nilai Intensitas')
plt.ylabel('Jumlah Piksel')
plt.xlim([0, 255])

plt.tight_layout()
plt.show()
```

kode ini berfungsi untuk menormalkan intensitas piksel ke rentang 0–255 guna memperbaiki kontras citra. Menggunakan rumus min-max, kode ini juga menangani kondisi khusus di mana jika input memiliki nilai yang seragam, output akan diseragamkan ke nilai 128. Analisis hasil dilakukan dengan membandingkan grafik histogram (diagram batang) antara data asli dan data yang telah dinormalisasi.

### 8. konvolusi

```python
kernel = np.array([
    [-1/10, -1/10, -1/10, -1/10],
    [-1/10,  6/10,  6/10, -1/10],
    [-1/10,  6/10,  6/10, -1/10],
    [-1/10, -1/10, -1/10, -1/10]
])

def conv2d_manual(image, kernel):
    h, w = image.shape
    kh, kw = kernel.shape
    pad_h, pad_w = kh//2, kw//2
    
    padded = np.zeros((h + 2*pad_h, w + 2*pad_w))
    padded[pad_h:pad_h+h, pad_w:pad_w+w] = image
    
    result = np.zeros_like(image, dtype=float)
    
    for i in range(h):
        for j in range(w):
            region = padded[i:i+kh, j:j+kw]
            result[i, j] = np.sum(region * kernel)
    
    result = np.clip(result, 0, 255).astype(np.uint8)
    return result

conv_result = conv2d_manual(kanvas_normalized, kernel)

plt.figure(figsize=(6, 6))
plt.imshow(conv_result, cmap='gray')
plt.title('Hasil Konvolusi')
plt.axis('on')
plt.tight_layout()
plt.show()
```

Bagian kode ini berfungsi untuk menerapkan "filter" khusus pada gambar agar menonjolkan fitur tertentu. Bayangkan kernel sebagai sebuah jendela kecil yang bergeser di seluruh permukaan gambar. Di setiap pergeserannya, jendela ini menghitung nilai baru untuk setiap titik piksel. Karena hasil hitungannya bisa sangat besar atau negatif, digunakan perintah np.clip supaya nilainya kembali normal di angka 0 sampai 255.
